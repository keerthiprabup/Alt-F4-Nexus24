```python
from datasets import load_dataset

dataset = load_dataset('yuncongli/chat-sentiment-analysis')

def format_sample(example):
    text = example['input']
    text2 = example['output']
    formatted_text = f'<s>[INST] {text} [/INST] {text2} </s>'
    return {'text': formatted_text}

transformed_dataset = dataset.map(format_sample)




```

```python
transformed_dataset.push_to_hub("finetune")
```
